```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Dataset': ['Diabetes', 'Diabetes', 
                'Credit', 'Credit', 
                'SBA', 'SBA', 
                'no2', 'no2'],

    'Method': ['mean Δ using MILP', 'mean Δ using our'] * 4,

    'Mean Δ': [0.108742,0.274099, 
               0.069975,0.203909,  
               	0.10966,0.317258, 
               0.022575,0.076827],
})

sns.set_palette("pastel")
# Plot the results with seaborn
plt.figure(figsize=(12, 8))
ax =sns.barplot(x='Dataset', y='Mean Δ', hue='Method', data=df, ci='sd', estimator="sum", capsize=0.1, errwidth=1.5, dodge=True, width=0.5)

plt.xlabel('Dataset')
plt.ylabel('Mean Δ')

ax.set_xlabel('Dataset', fontsize=20)
ax.set_ylabel('Mean Δ', fontsize=20)
ax.tick_params(axis='both', labelsize=20)
ax.legend(fontsize=15)


plt.savefig('plot.pdf')
plt.show()
```
